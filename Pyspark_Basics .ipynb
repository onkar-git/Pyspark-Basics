# Py Spark ML

Dataframe API framework

```python
from pyspark.sql import SparkSession
spark =SparkSession.builder.appName('ml').getOrCreate()
```

```python
dataset=spark.read.csv("data.csv",header=True,inferSchema=True)
```

```python
dataset.show()
```

```python
dataset.columns
```

```python
# we have to basicaly create spiltting data here differently, we use the vector assemble
```

```python
#[age,exp]-------> this will be my new independant feature
```

```python
from pyspark.ml.feature import VectorAssembler
feature_assembler = VectorAssembler(inputCols=['age','exp'],outputCol="Independent_features")
```

```python
output = feature_assembler.transform(dataset)
```

```python
output.show()
```

```python
output.columns
```

```python
final_data=output.select("Independent_features","salary")
```

```python
final_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
train_data,test_data=final_data.randomSplit([0.8,0.2])
regressior=LinearRegression(featuresCol='Independent_features', labelCol='salary')
regressior=regressior.fit(train_data)
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
regressior.coefficients
```

```python
regressior.intercept
```

```python
#prediction
pred_results=regressior.evaluate(test_data)
```

```python
pred_results.predictions.show()
```

```python
pred_results.meanAbsoluteError,pred_results.meanSquaredError
```
